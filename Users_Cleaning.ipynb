# Cleaning

```python
users_relevant = ['createdAt', 'gender', 'interests' , 'is_verified','status','updatedAt','who_to_date']
```

```python
users_df = users[users_relevant]
```

```python
users_df
```

```python
users_df["interests"][0]
```

```python
users_df.isnull().sum()
```

```python
users_df['status'].sum()
```

```python
users_df  = users_df.drop(users_df[users_df.status==False].index)
```

```python
users_df.info()
```

```python
users_df.isnull().sum()
```

```python
users_df  = users_df [users_df ['gender'].notna()]
```

```python
users_df.isnull().sum()
```

```python
users_df.info()
```

```python
users_df['who_to_date'].fillna("A", inplace = True)
 
```

```python
users_df.info()
```

```python
users_df.head()
```

```python
def format_interests(s):
  return re.sub("\"|\[|\]", "", s)

users_df['interests'] = users_df.interests.apply(lambda x: format_interests(str(x))) 
```

```python
users_df["createdAtTime"]=users_df['createdAt']

users_df["updatedAtTime"]=users_df['updatedAt']
```

```python
users_df['createdAt'] = pd.to_datetime(users_df['createdAt'])
# users_df['createdAtTime'] = pd.to_datetime(users_df.createdAtTime).dt.strftime('%H:%M:%S')

users_df['updatedAt'] = pd.to_datetime(users_df['updatedAt'])
# users_df['updatedAtTime'] = pd.to_datetime(users_df.updatedAtTime).dt.strftime('%H:%M:%S')
```
